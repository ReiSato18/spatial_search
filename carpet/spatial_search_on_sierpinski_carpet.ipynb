```julia
n_4 = 3^4; n_5 = 3^5; n_6 = 3^6

function cp(n, cp_stage)
    carpet = zeros(Int64, n+2, n+2)
    for i in 1:n+2
        carpet[1,i] = 2
        carpet[n+2,i] =2
        carpet[i,1] = 2
        carpet[i,n+2] = 2
    end
    carpet[2:n+1, 2:n+1] = cp_stage
    return carpet
end
cp(n_4, stage4)
```

```julia
C = [-1/2 1/2 1/2 1/2;1/2 -1/2 1/2 1/2;1/2 1/2 -1/2 1/2;1/2 1/2 1/2 -1/2]
```

```julia
function coin(psi)
    csi = C*psi
    return csi
end
```

```julia
using LinearAlgebra
using ProgressMeter
using Plots

function walk(s,n,node,mark,step, carpet)
    mark_prob=Float64[]
    
    stage = zeros(Float64,n,n,4)
    stage[mark,mark,:] = [1 1 -1 -1]/2
    for x=1:n, y=1:n
        if carpet[x,y]==1
            stage[x,y,:] = [1/sqrt(4node),1/sqrt(4node),1/sqrt(4node),1/sqrt(4node)]
        else
            continue
        end
    end
    progress=Progress(step)
    #anim = @animate for t in 0:step
    for t in 0:step
        stage[mark,mark,:] *= -1
        next_stage = zeros(n,n,4)
        for x =1:n,y=1:n
            x0 = ((x-1 + (n-1))%n) + 1
            x1 = ((x+1 + (n-1))%n) + 1
            y0 = ((y-1 + (n-1))%n) + 1
            y1 = ((y+1 + (n-1))%n) + 1
            
        
            psi = coin(stage[x,y,:])
            #flip_flop
            if carpet[x,y]==1 && carpet[x0,y] == 1 
                next_stage[x0,y,1] = psi[2] 
            elseif carpet[x,y]==1 && carpet[x0,y] == 0
                next_stage[x,y,2] = psi[2]
            else
                continue
            end
            ####
            if carpet[x,y]==1 && carpet[x1,y] == 1
                next_stage[x1,y,2]=psi[1]
            elseif carpet[x,y]==1 && carpet[x1,y] == 0
                next_stage[x,y,1]=psi[1]
            else
                continue
            end
            ####
            if carpet[x,y]==1 && carpet[x,y0] == 1
                next_stage[x,y0,3]=psi[4]
            elseif carpet[x,y]==1 && carpet[x,y0] == 0
                next_stage[x,y,4]=psi[4]
            else
                continue
            end
            ####
            if carpet[x,y]==1 && carpet[x,y1]==1
                next_stage[x,y1,4]=psi[3]
            elseif carpet[x,y]==1 && carpet[x,y1]==0
                next_stage[x,y,3]=psi[3]
            else
                continue
            end
        end
        stage = copy(next_stage)
        next!(progress)
        
        
        #one_side = n
        #hist = zeros(Float64,n,n)
        #for x in 1:n, y in 1:n
        #    if carpet[x,y] == 0
        #        hist[x,y] = 0#NaN
        #    elseif carpet[x,y] == 2
        #        hist[x,y] = 0
        #    else
        #        hist[x,y] = copy(dot(stage[x,y,:], stage[x,y,:]))
        #    end
        # end
         #Plots.plot(1-sum(hist))
         #push!(mark_prob, 1-sum(hist))
         #Plots.heatmap(hist, aspectratio=1)#, clim=(0,0.3) ) 
         #push!(mark_prob,sum(hist[mark,mark,:]) )
        push!(mark_prob, dot(stage[mark,mark,:],stage[mark,mark,:]) )
    end
    #return stage
    #return gif(anim,fps=20)
    #return stage
    return mark_prob
end
```
